```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_spectrum(ax, mc_fname, data_fmt='o', 
                  title=None, xlabel=None, ylabel=None):
    
    mc_arr = np.genfromtxt(mc_fname)
    cts, edges = np.histogram(mc_arr, bins='auto')
    y, yerr = cts, np.sqrt(cts)
    x = edges[:-1]+(edges[1:]-edges[:-1])/2

    ax.errorbar(x, y, yerr=yerr, fmt=data_fmt)
    
    if title:
        ax.set_title(title)
    if xlabel:
        ax.set_xlabel(xlabel)
    if ylabel:
        ax.set_ylabel(ylabel)
```

```python
fig = plt.figure(figsize=(10*1.618,20))
ax1 = fig.add_subplot(4,2,1)
plot_spectrum(ax1, 'D0ss0.csv', title=r'$D^{0}_0$')
ax2 = fig.add_subplot(4,2,2)
plot_spectrum(ax2, 'D0ssc.csv', title=r'$D^{+}_0$')
ax3 = fig.add_subplot(4,2,3)
plot_spectrum(ax3, 'D10.csv', title=r'$D^{0}_1$')
ax4 = fig.add_subplot(4,2,4)
plot_spectrum(ax4, 'D1c.csv', title=r'$D^{+}_1$')
ax1 = fig.add_subplot(4,2,5)
plot_spectrum(ax1, 'D20.csv', title=r'$D^{0}_2$')
ax2 = fig.add_subplot(4,2,6)
plot_spectrum(ax2, 'D2c.csv', title=r'$D^{+}_2$')
ax3 = fig.add_subplot(4,2,7)
plot_spectrum(ax3, 'D10.csv', title=r"$D^{'}_1$")
ax4 = fig.add_subplot(4,2,8)
plot_spectrum(ax4, 'D1c.csv', title=r"$D^{'+}_1$")
```
