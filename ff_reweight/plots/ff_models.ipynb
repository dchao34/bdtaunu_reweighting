```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_ff(ax, mc_fname, ff_fname, 
            data_fmt='o', ff_fmt='-'):
    
    mc_arr = np.genfromtxt(mc_fname)
    ff_arr = np.genfromtxt(ff_fname)
    
    cts, edges = np.histogram(mc_arr, bins='auto')
    w = np.mean(edges[1:]-edges[:-1])
    density, density_edges = np.histogram(mc_arr, bins='auto', density=True)
    y, yerr = density, np.sqrt(cts*(sum(cts)-cts)/(float(sum(cts)))**3/w**2)
    x = edges[:-1]+(edges[1:]-edges[:-1])/2
    
    ax.errorbar(x, y, yerr=yerr, fmt=data_fmt)
    ax.plot(ff_arr[:,0], ff_arr[:,1], ff_fmt)
```

```python
fig = plt.figure(figsize=(10*1.618,10))
ax1 = fig.add_subplot(2,2,1)
plot_ff(ax1, 'BDe.csv', 'BDenu_ISGW2.csv')
ax2 = fig.add_subplot(2,2,2)
plot_ff(ax2, 'BDmu.csv', 'BDmunu_ISGW2.csv')
ax3 = fig.add_subplot(2,2,3)
plot_ff(ax3, 'BDtau.csv', 'BDtaunu_ISGW2.csv')
```
