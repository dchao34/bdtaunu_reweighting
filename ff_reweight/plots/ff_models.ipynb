```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
save_pdf = True
```

```python
def plot_ff(ax, mc_fname, ff_fname, lw=2, 
            title=None, xlabel=None, ylabel='density',
            data_fmt='ko', ff_fmt='-'):
    
    mc_arr = np.genfromtxt(mc_fname)
    ff_arr = np.genfromtxt(ff_fname)
    
    cts, edges = np.histogram(mc_arr, bins='auto')
    w = np.mean(edges[1:]-edges[:-1])
    density, density_edges = np.histogram(mc_arr, bins='auto', density=True)
    y, yerr = density, np.sqrt(cts*(sum(cts)-cts)/(float(sum(cts)))**3/w**2)
    x = edges[:-1]+(edges[1:]-edges[:-1])/2
    
    ax.errorbar(x, y, yerr=yerr, fmt=data_fmt)
    ax.plot(ff_arr[:,0], ff_arr[:,1], ff_fmt, lw=lw)
    
    if title: ax.set_title(title)
    if xlabel: ax.set_xlabel(xlabel)
    if ylabel: ax.set_ylabel(ylabel)

mpl.rcParams.update({'font.size': 20});
```

```python
fig = plt.figure(figsize=(6*1.618,6))
ax = fig.add_subplot(1,1,1)
plot_ff(ax, 'BDe.csv', 'BDenu_ISGW2.csv', 
        title=r'$B\rightarrow D e \nu$', xlabel=r'$q^2$', ylabel=r'$d\Gamma/dq^2$ (density)')

if save_pdf: fig.savefig('BDenu_ff_sim.pdf')
```
