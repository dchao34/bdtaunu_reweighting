```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_spectrum(ax, mc_fname, 
                  data_fmt='o'):
    
    mc_arr = np.genfromtxt(mc_fname)
    
    cts, edges = np.histogram(mc_arr, bins='auto')
    y, yerr = cts, np.sqrt(cts)
    x = edges[:-1]+(edges[1:]-edges[:-1])/2

    ax.errorbar(x, y, yerr=yerr, fmt=data_fmt)
```

```python
fig = plt.figure(figsize=(10*1.618,10))
ax1 = fig.add_subplot(2,2,1)
plot_spectrum(ax1, 'BDe.csv')
ax2 = fig.add_subplot(2,2,2)
plot_spectrum(ax2, 'BDmu.csv')
ax3 = fig.add_subplot(2,2,3)
plot_spectrum(ax3, 'BDtau.csv')
```
