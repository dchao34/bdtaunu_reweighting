```python
import numpy as np
import matplotlib.pyplot as plt
import iopro
```

```python
arr = np.genfromtxt('weight_visual.csv', names=True)
```

```python
def apply_cut(arr, varname, lower=None, upper=None):
    sub_arr = np.copy(arr)
    mask = np.ones(sub_arr.shape[0], dtype=bool)
    if lower is not None: mask &= (sub_arr[varname] >= lower)
    if upper is not None: mask &= (sub_arr[varname] <= upper)
    sub_arr = sub_arr[mask]
    return sub_arr
```

```python
X = apply_cut(arr, 'z1', lower=-4)
X = apply_cut(X, 'z2', lower=-5, upper=3)
```

```python
fig = plt.figure(figsize=(1.618*6, 6))
ax = fig.add_subplot(1,1,1)
ax.hist(X['z1'], bins=200, weights=X['w'], histtype='stepfilled', alpha=0.5);
ax.hist(X['z1'], bins=200, weights=X['sideband_w']*X['w'], histtype='stepfilled', alpha=0.5);
```

```python
fig = plt.figure(figsize=(1.618*6, 6))
ax = fig.add_subplot(1,1,1)
ax.hist(X['z2'], bins=100, weights=X['w'], histtype='stepfilled', alpha=0.5);
ax.hist(X['z2'], bins=100, weights=X['sideband_w']*X['w'], histtype='stepfilled', alpha=0.5);
```
