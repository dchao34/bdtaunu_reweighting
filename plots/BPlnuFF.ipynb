```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
arr1 = np.genfromtxt('test8_isgw2.csv')
arr2 = np.genfromtxt('test8_cln.csv')
```

```python
mpl.rcParams.update({'font.size': 20});

fig = plt.figure(figsize=(6*1.618, 6));
ax = fig.add_subplot(1,1,1);
ax.plot(arr1[:,0], arr1[:,1], '-', label='ISGW2');
ax.plot(arr2[:,0], arr2[:,1], '-', label='CLN');
ax.set_xlabel(r'$q^2$ (GeV)');
ax.set_ylabel(r'density');
plt.legend(fontsize=16);
fig.suptitle(r'$d\Gamma/dq^2$ for $B\rightarrow D\tau\nu$');
```
