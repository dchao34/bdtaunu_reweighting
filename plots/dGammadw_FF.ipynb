```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
arr1 = np.genfromtxt('BDss0enu_ISGW2.csv')
arr2 = np.genfromtxt('BDss0enu_LLSWB1.csv')
arr3 = np.genfromtxt('BDss0enu_LLSWB2.csv')

mpl.rcParams.update({'font.size': 20});

fig = plt.figure(figsize=(6*1.618, 6));
ax = fig.add_subplot(1,1,1);
ax.plot(arr1[:,0], arr1[:,1], '-', label='ISGW2');
ax.plot(arr2[:,0], arr2[:,1], '-', label='LLSWB1');
ax.plot(arr3[:,0], arr3[:,1], '-', label='LLSWB2');
ax.set_xlabel(r'$w$');
ax.set_ylabel(r'density');
plt.legend(fontsize=16, loc=2);
fig.suptitle(r'$d\Gamma/dw$ for $B\rightarrow D^{*}_0e\nu$');
```
