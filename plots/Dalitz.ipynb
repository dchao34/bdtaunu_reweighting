```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
arr = np.genfromtxt('test7.csv')
```

```python
mpl.rcParams.update({'font.size': 20})

fig = plt.figure(figsize=(6*1.618, 6));
ax = fig.add_subplot(1,1,1);
ax.plot(arr[:,1], arr[:,0], '.');
ax.set_xlabel(r'$E_\ell$ (GeV)');
ax.set_ylabel(r'$q^2$ (GeV$^2$)');
fig.suptitle(r'$B\rightarrow D^*e\nu$ Dalitz Plot with ISGW2');

props = dict(boxstyle='round', color='yellow', alpha=0.3)
ax.text(0.05, 0.95, 'Stray points due to FSR', transform=ax.transAxes,
        verticalalignment='top', fontsize=18, bbox=props);
```
